# Funções


```python
#ESCOPO DE VARIÁVEIS

#VARIAVEL LOCAL
def funcaoLocal():
    variavelLocal = 10
    print(variavelLocal)


funcaoLocal()
#print(variavelLocal) Gera erro
```

```python
#ESCOPO DE VAIRÁVEIS

#VARIÁVEL GLOBAL

variavelGlobal = 5

def funcaoGlobal():
    global variavelGlobal
    variavelGlobal += 1
    print(variavelGlobal)


print(variavelGlobal)
funcaoGlobal()
print(variavelGlobal)
```

```python
#FUNÇÃO SEM PARÂMETRO

def saudacao():
    print("Hello, World!!")

saudacao()
```

```python
#FUNÇÃO COM PARÃMETROS (OU ARGUMENTO)

def soma(x, y):
    resultado = x + y
    print(resultado)

soma(2, 4)
```

```python
#FUNÇÃO COM PARÃMETRO E RETORNO

def soma(x, y):
    resultado = x + y
    return resultado

print(soma(2, 4) - soma(1, 1))
```

```python
#FUNÇÃO COM VALOR PADRÃO

def divisao(x = 5, y = 2):
    resultado = x / y
    return resultado

print(divisao(10))
```

```python
#EMPACOTAMENTO DE DADOS

def mult(*numeros):
    calculo = 1
    for i in numeros:
        calculo*=i
    return calculo

print(mult(2, 2, 2))
```

```python
#EXEMPLO

num = (2,4,6,8)

def mult(numeros):
    calculo = 1
    for i in numeros:
        calculo*=i
    return calculo

print(mult(num))
```

```python
# DESAFIO 01
"""
Crie uma função que calcule a soma de diversos números.
"""

def soma(*numeros):
    calculo = 0
    for i in numeros:
        calculo += i
    return calculo

print(som(4,6,7,9,10))
```

```python
# DESAFIO 02
"""
Faça um programa que receba um número inteiro de 0 a 9.
Crie uma função que apresente o resultado da tabuada deste número.
"""

def tabuada(inteiro):
    for i in range(1, 11):
        print(f"{i} x {inteiro} = {inteiro * i}")

n = int(input("insira um número: "))
tabuada(n)
```

```python
# DESAFIO 03
"""
Faça um programa que tenha uma função chamada área(), que receba as
dimensões de um terreno retangular (largura e comprimento) pelo usuário e mostre a área do terreno.
"""

def area(l, c):
    return l * c

l = int(input("Insira a largura: "))
c = int(input("Insira o comprimento: "))

print(area(l, c))
```
