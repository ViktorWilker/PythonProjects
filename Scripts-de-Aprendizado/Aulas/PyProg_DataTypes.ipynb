```python
# TIPOS NUMÉRICOS (int float)

x  = 10     # Inteiro
y  = -1     # Inteiro negativo
z  = -5.29  # Decimal negativo
pi = 3.14   # Decimal

print(type(x))
print(type(y))
print(type(z))
print(type(pi))
```

```python
# TIPOS DE DADOS CARACTERES (str)

nome = 'Leandro'
escola = "SENAI Ary Torres"
print(nome)
print(escola)
print(type(nome))
print('\n',escola[0:5])
```

```python
# TIPOS LÓGICOS (bool)

continuar = True
parar     = False
casado    = True

print(continuar)
print(parar)
print(type(casado))
```

```python
# DECLARAÇÃO DE VARIÁVEIS

texto      = '2'
texto1     = str(2)
numero     = 3
numero1    = int(3)
decimal    = float(3)
logico     = bool(False)

print(texto)
print(texto1)
print(numero)
print(numero1)
print(decimal)
print(logico)
```

```python
# ENTRADA DE DADOS (input)

numeroInteiro = int(input("Digite um número inteiro: "))
print(numeroInteiro)

numeroDecimal = float(input("\nDigite um número decimal: ")) # Sepador de decimal: . (não vírgula)
print(numeroDecimal)

valorLogico = bool(input("\nDigite algo: "))
print(valorLogico)

# Não há necessidade declarar str() para o input()
texto = str(input("\nDigite um valor alfanumérico:\n"))
print(texto)
```

```python
# IDENTIFICAR TIPO PRIMITIVO

print(type(numeroInteiro))
print(type(numeroDecimal))
print(type(valorLogico))
print(type(texto))
```

```python
# ANÁLISE DE STRINGS (str)

n = '5t'

print(n.isnumeric())   # Númérico
print(n.isalpha())     # Alfabético
print(n.isalnum())     # Alfanumérico
print(n.isupper())     # Maiúscula
print(n.islower())     # Minúscula
print(n.istitle())     # Primeira maiúscula
print(n.isdecimal())   # Numérico
print(n.isspace())     # Espaço
```
