```python
# EXEMPLO 04

for i in range(5):
    print(f"{i+1} | SENAI")
```

```python
# EXEMPLO 05

texto = "PYTHON"

for letra in texto:
    print(letra, end = ' ')
```

```python
# EXEMPLO 06

frase = ['PYTHON', 'É', 'FANTÁSTICO']

for palavra in frase:
    print(palavra)
```

```python
# EXEMPLO 07
""" continue: pula o resto da iteração"""

lista = [1, 10, 20, 30, 40, 50]

for numero in lista:
    if numero == 30:
        continue
    print(numero, end = ' ')
```

```python
# EXEMPLO 08
""" """

for i in range(50):
    if i >= 10:
        break
    print(i, end = ' ')
```

```python
# EXEMPLO 09

for i in range(1, 11):
    print()
    for j in range(1, 11):
        print(f"{i*j}".rjust(3), end=' ')
```

```python
# DESAFIO 02
"""
Faça um programa que calcule a soma entre todos os números ímpares
que são múltiplos de três e que se encontram no intervalo de 1 até 500.
"""
n = 0

for i in range(1, 500):
    if i % 2 != 0 and i % 3 == 0:
        n += i


print(n)
```

```python
# DESAFIO 07
"""
Faça um programa que leia o peso de cinco pessoas.
No final, mostre qual foi o maior e o menor peso lidos.
"""

pesoMaior = 0
pesoMenor = float('inf')

for i in range(4):
    currentPeso = float(input(f"Insira o peso da {i+1}° pessoa: "))
    if currentPeso > pesoMaior:
        pesoMaior = currentPeso

    if currentPeso < pesoMenor:
        pesoMenor = currentPeso


print(pesoMaior)
print(pesoMenor)
```

```python
# DESAFIO 08
"""
Crie um programa que leia o ano de nascimento de sete pessoas. No final,
mostre quantas pessoas ainda não atingiram a maioridade e quantas já são maiores.
"""
ano = 2024
maiorIdade = 18
x = 0
y = 0

for i in range(6):
    currentYear = int(input(f"Insira o ano de nascimento da {i}° pessoa: "))

    if ano - currentYear >= maiorIdade:
        x += 1
    else:
        y += 1


print(f"Quantidade de pessoas que atingiram a maioridade: {x}")
print(f"Pessoas que não atigiram: {y}")
```

```python
# DESAFIO 10
"""
Faça um programa que leia um número inteiro e diga se ele é ou não um numero primo.
>1 só é divisível por ele mesmo
"""

n = int(input("Insira um número maior que 1: "))
x = 0

for i in range(1, n):
    if n % i == 0:
        x += 1
    if x > 1:
        break

if x == 1:
    print("é número primo")
else:
    print("Não é número primo")
```
