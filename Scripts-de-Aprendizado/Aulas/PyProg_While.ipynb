```python
# DESAFIO 02
"""
Fa√ßa um programa onde o computador vai ‚Äúpensar‚Äù em um n√∫mero entre 1 a 10.
O jogador vai tentar adivinhar at√© acertar, mostrando no final quantos
palpites foram necess√°rios para vencer.
"""

from google.colab import output
import random
import time

r = random.randrange(1, 10)
palpites = 0
n = 0

while True:
    n = int(input("O Computador sorteou um n√∫mero entre 1 a 10, acerte qual √© esse n√∫mero: \n"))

    if n == r:
        print("Parab√©ns! Voc√™ acertou o n√∫mero")
        break
    else:
        palpites += 1
        print("Voc√™ errou! Continue tentando!")

    time.sleep(2)
    output.clear()

print(f"Tentativas at√© o acerto: {palpites}")
```

```python
# DESAFIO 04
"""
Fa√ßa um programa que leia um n√∫mero qualquer e mostre o seu
fatorial.
Exemplo:

5! = 5 x 4 x 3 x 2 x 1 = 120
"""

n = int(input("Insira um n√∫mero: "))
i = 1
x = 1

while i <= n:
    x *= i
    i+=1

print(f"{n}! = {x}")
```

```python
# DESAFIO 07
"""
Crie um programa que leia v√°rios n√∫meros inteiros pelo teclado. O
programa s√≥ vai parar quando o usu√°rio digitar o valor 999, que √© a
condi√ß√£o de parada. No final, mostre quantos n√∫meros foram
digitados e qual foi a soma entre eles (desconsiderando o 999).
"""

x = 0
y = 0

while True:
    n = int(input("Digite um n√∫mero: "))
    if n >= 999:
        break

    x += n
    y += 1


print(f"\nN√∫meros digitados: {y}")
print(f"Soma dos n√∫meros: {x}")
```

```python
# DESAFIO 08
"""
Crie um programa que leia v√°rios n√∫meros inteiros pelo teclado. No
final da execu√ß√£o, mostre a m√©dia entre todos os valores e qual foi o
maior e o menor valores lidos. O programa deve perguntar ao usu√°rio
se ele quer ou n√£o continuar a digitar valores
"""
from google.colab import output

m = 0
n = 0
maior = 0
menor = float('inf')

while True:
    x = int(input("\nDigite um n√∫mero inteiro: "))
    n += 1
    m += x

    if x > maior:
        maior = x

    if x < menor:
        menor = x

    opc = input("Gostaria de inserir outro valor?(S/N)\n").upper()
    if opc == "N":
        output.clear()
        break
    else:
        output.clear()
        continue

print(f"\nM√©dia: {m/n}")
print(f"Maior: {maior}")
print(f"Menor: {menor}")
```

```python
# DESAFIO 09
"""
Escreva um programa que leia um n√∫mero n inteiro qualquer e mostre
na tela os n primeiros elementos de uma sequ√™ncia de Fibonacci.
Exemplo:

0 ‚Äì> 1 ‚Äì> 1 ‚Äì> 2 ‚Äì> 3 ‚Äì> 5 ‚Äì> 8
"""

x = 1
y = 0
z = 0
i = 0

n = int(input("Insira quantos elementos ter√° a sequ√™ncia(min: 3): "))

while i <= n:
    print(y, end = ' ')
    y = x + z
    z = x
    x = y

    i+=1
```
