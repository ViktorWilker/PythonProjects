```python
#MANIPULAÇÃO DE LISTAS

bancos = ["Banco do Brasil", "CEF", "Santander"]
print(bancos)
print(bancos[2])

#SUBSTITUIR ITEM DA LISTA
bancos[1] = "Itaú"
print(bancos)

#SUBSTITUIR O ÚLTIMO ITEM DA LISTA
bancos[-1] = "C6"
print(bancos)

#AGREGAR ITENS A LISTA(operador de atribuição)
bancos += ["Bradesco", "Nubank"]
print(bancos)

#AGREGAR ITENS A LISTA
bancos = bancos + ["Safra"]
print(bancos)

```

```python
#MÉTODOS PARA LISTAS

lista = [4, 5, 3, 5]
print(lista)

#ADICIONAR ITENS A LISTA
lista.append(2)
print(lista)

#ADICIONAR VALOR NO ÍNDICE ESPECÍFICO
lista.insert(2, -3)
print(lista)

#CONTAR VALORES
print(lista.count(5))

#VERIFICAR TAMANHO DA LISTA
print(len(lista))

#OBTER ÍNDICE DO ITEM
print(lista.index(5))

#REVERTER ORDEM DA LISTA
lista.reverse()
print(lista)

#ORDERNAR LISTA
lista.sort()
print(lista)

#REMOVER ITEM ATRAVÉS DO VALOR
lista.remove(4)
print(lista)

#REMOVER O ÚLTIMO ITEM DA LISTA (recebe o valor removido)
removeu = lista.pop()
print(removeu)
print(lista)

#REMOVER INTERVALO (ou índice)
del lista[1:3]
print(lista)

#LIMPAR A LISTA
lista.clear()
print(lista)

```

```python
# DESAFIO 01
"""
Faça um programa que leia 5 valores numéricos e
guarde-os em uma lista. No final, mostre qual foi
o maior e o menor valor digitado e as suas respectivas posições na lista.
"""

n = []

for i in range(5):
    x = int(input("Insira um número: "))
    n.append(x)

print(f"\nMaior número: {max(n)}, Respectivo índice: {n.index(max(n))}")
print(f"Menor número: {min(n)}, Respectivo índice: {n.index(min(n))}")
```

```python
# DESAFIO 02
"""
Crie um programa onde o usuário possa digitar vários valores
numéricos e cadastre-os em uma lista. Caso o número já exista
lá dentro, ele não será adicionado. No final serão exibidos
todos os valores únicos digitados, em ordem crescente.
"""

n = int(input("Quantos números deseja inserir? "))
lista = []

for i in range(n):
    x = int(input("\nInsira um número: "))
    if x not in lista:
        lista.append(x)


lista.sort()
print("\n",lista)
```

```python
# DESAFIO 03
"""
Crie um programa que vai ler vários números e colocar em uma lista.
Depois disso, mostre:
A) Quantos números foram digitados.
B) A lista de valores, ordenada de forma decrescente.
C) Se o valor 5 foi digitado e esta ou não na lista.
"""

n = int(input("Quantos números deseja inserir? "))
lista = []

for i in range(n):
    x = int(input("\nInsira um número: "))
    lista.append(x)

lista.sort()
lista.reverse()
print(f"\nNúmeros digitados: {len(lista)}")
print(f"Valores em ordem decrescente: ", lista)
print("5 está na lista" if 5 in lista else "5 não está na lista")
```
