```python
# DESAFIO 01
"""
Exiba apenas a linha que possui o Total
"""

arquivo = "dadosOrcamento.txt"

with open(f'{path}{arquivo}', 'r') as arquivo:
    texto = arquivo.readlines()
    for linha in texto:
        if "Total" in linha:
            print(linha)
```

```python
# DESAFIO 02
"""
Crie um sistema de login que faça a leitura do arquivo login.txt que está na seguinte disposição:
root
1234
Leandro Toniati
Caso digite o login(root) e senha(1234) buscando as informações no arquivo, será apresentada
uma mensagem de boas vindas com o nome do usuário, caso contrário informe 'Usuário ou senha não conferem'
Utilize para a senha:
Biblioteca: import getpass
Método: getpass.getpass("Senha: ")
"""

import getpass

arquivo = "login.txt"

with open(f"{path}{arquivo}", 'r') as arquivo:
    login = input("Login: ")
    senha = getpass.getpass("Senha: ")
    datas = arquivo.readlines()

    for i in range(len(datas)):
        if "root" in datas[0]:
            if "1234" in datas[1]:
                print(f"Bem vindo {datas[2]}")
                break
```

```python
# DESAFIO 03
"""
Melhore o DESAFIO 02 limitando o login a 3 tentativas. Caso o usuário erre o Login ou Senha
pela 3ª vez exiba a mensagem: "Número de tentativas excedido"
"""

import getpass
i = 3

arquivo = "login.txt"

with open(f"{path}{arquivo}", 'r') as arquivo:
    while i > 0:
        login = input("Login: ")
        senha = getpass.getpass("Senha: ")
        datas = arquivo.readlines()
        for j in range(len(datas)):
            if "root" in datas[0]:
                if "1234" in datas[1]:
                    print(f"Bem vindo {datas[2]}")
                    break
                else:
                    i -= 1
                    print(f"Você possui {i} tentativas")
            else:
                i -= 1
                print(f"Você possui {i} tentativas")
```
