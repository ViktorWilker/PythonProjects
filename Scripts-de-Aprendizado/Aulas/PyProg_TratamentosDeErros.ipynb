# Tratamento de Erros

```python
#EXEMPLO 01

try:
    #código pode gerar erro...
    numero = int(input("Digite um número: "))
    resultado = 10/numero
except(ValueError, TypeError):
    #Trata os erros mencionados acima
    print("Valor ou tipo inválidos...")
except(ZeroDivisionError):
    #Trata erro mencionado acima
    print("Não é possível dividir por zero...")
else:
    #Executa caso não houver erros
    print(f"Resultado: {resultado}")
finally:
    #Sempre executa
    print("\nEncerrando o programa")
```

```python
#EXEMPLO 02

try:
    numerador = int(input("Digite o numerador: "))
    denominador = int(input("Digite o denominador: "))
    razao = numerador/denominador
except Exception as erro:
    print(f"Erro encontrado: {erro}")
    print(f"Erro encontrado: {erro.__class__}")
else:
    print(razao)
```

```python
# DESAFIO 01

try:
    valorImovel        = float(input("Digite o valor do imóvel: "))
    salario            = float(input("Digite seu salário: "))
    mesesFinanciamento = int(input("Digite prazo de pagamento em anos: "))*12

    valorParcela = valorImovel/mesesFinanciamento
    salarioTrintaPorcento = salario*0.3
except(ValueError, TypeError):
    print("Informe valores válidos...")
except(ZeroDivisionError):
    print("O ano não pode ser zero...")
else:
    if valorParcela > salarioTrintaPorcento:
        print(f"\n\033[31mNegado!\033[0m Parcela máxima: R$ {salarioTrintaPorcento:,.2f}")
        print(f"Parcela simulada: R$ {valorParcela:,.2f}")
    else:
        print(f"\n\033[32mAprovado!\033[0m Valor da parcela: R$ {valorParcela:,.2f}")
```
