# Match Case

```python
# OPÇÃO ÚNICA

operacao = input("""
Escolha uma opção:

[1] Opção 1
[2] Opção 2
[3] Opção 3
""")

match operacao:
    case '1':
        print("Opção 1 foi escolhida")
    case '2':
        print("Opção 2 foi escolhida")
    case '3':
        print("Opção 3 foi escolhida")
    case _:
        print("Opção inválida")
```

```python
# OPÇÃO MÚLTIPLA

operacao = input("""
Escolha uma opção:

[1] Opção 1
[2] Opção 2
[3] Opção 3
""")

match operacao:
    case '1' | '2':
        print("Opção 1 OU 2 escolhida")
    case '3':
        print("Opção 3 foi escolhida")
    case _:
        print("Opção inválida")
```

```python
# OPÇÃO MÚLTIPLA COM case _ if

numero = int(input("Digite um número: "))

match numero:
    case 1 | 3 | 5 | 7 | 9:
        print("É ímpar entre 1 e 10!")
    case _ if numero < 0:
        print("É um número negativo!")
    case _:
        print("Um número qualquer...")
```

```python
# DESAFIO 01
"""
Escreva um programa em Python que solicite ao usuário uma nota inteira entre 0 e 10.
Use o comando `match case` para classificar a nota de acordo com a tabela a seguir:

- 0-4: Nota Baixa
- 5-7: Nota Média
- 8-9: Nota Alta
- 10: Nota Excelente

Exiba a classificação correspondente.
"""

nota = int(input("Insira uma nota inteira de 0 a 10: "))

match nota:
    case 0 | 1 | 2 | 3 | 4:
        print("Nota Baixa")
    case 5 | 6 | 7:
        print("Nota média")
    case 8 | 9:
        print("Nota Alta")
    case 10:
        print("Nota excelente")
    case _ :
        print("Nota inválida")
```

```python
# DESAFIO 02
"""
Refaça o DESAFIO 01 considerando notas decimais.
"""

nota = float(input("Insira uma nota de 0 a 10: "))

match nota:
    case _ if nota <= 4:
        print("Nota Baixa")
    case _ if nota <= 7:
        print("Nota média")
    case _ if nota <= 9:
        print("Nota Alta")
    case _ if nota == 10:
        print("Nota excelente")
    case _ if nota < 0:
        print(f"\nNota inválida")
```

```python
# DESAFIO 03
"""
Crie um programa que peça ao usuário para inserir um número de 1 a 7,
onde cada número representa um dia da semana (1 para domingo, 2 para segunda-feira e assim por diante).
Use match case para imprimir o nome do dia da semana correspondente.
"""

n = int(input("Insira um número de 1 a 7: "))

match n:
    case 1:
        print("Domingo")
    case 2:
        print("Segunda-feira")
    case 3:
        print("Terça-feira")
    case 4:
        print("Quarta-feira")
    case 5:
        print("Quinta-feira")
    case 6:
        print("Sexta-feira")
    case 7:
        print("Sábado")
```

```python
# DESAFIO 04
"""
Implemente uma calculadora simples que realiza as operações de adição, subtração, multiplicação e divisão.
O usuário deve inserir dois números e, em seguida, selecionar a operação desejada (+, -, *, /).
Use o comando match case para realizar a operação correta e exibir o resultado.
"""

n1 = int(input("Insira o primeiro número: "))
operador = input("Insira o oporador (+,-,*,/): ")
n2 = int(input("Insira o segundo número: "))

match operador:
    case _ if operador == "+":
        print(n1 + n2)
    case _ if operador == "-":
        print(n1 - n2)
    case _ if operador == "*":
        print(n1 * n2)
    case _ if operador == "/":
        print(n1 / n2)
```

```python
# DESAFIO 05
"""
Escreva um programa que converta valores entre diferentes moedas.
O usuário deve inserir o valor em reais e selecionar a moeda para conversão:
D (Dólar), E (Euro), ou L (Libra). Utilize match case para aplicar a conversão
correta com base nas taxas fictícias fornecidas a seguir:

1 Real = 0.18 Dólar
1 Real = 0.16 Euro
1 Real = 0.13 Libra

Símbolos: $ € £
"""

n = float(input("Insira um valor em reais: "))
moeda = input(" D - (Dólar), E - (Euro), L - (Libra): ").upper()

match moeda:
    case _ if moeda == "D":
        print(f"O Valor em dólares é: {n * 0.18}$")
    case _ if moeda == "E":
        print(f"O valor em Euros é: {n * 0.16}€")
    case _ if moeda == "L":
        print(f"O valor em Libras é: {n * 0.13}£")
```

```python
# DESAFIO 06
"""
Crie um programa que peça ao usuário para inserir uma letra. O programa deve utilizar match case
para verificar se a letra é uma vogal.
"""

letra = input("Insira uma letra: ").lower()
vogais = "aeiou"

match letra:
    case _ if letra in vogais:
        print("é uma vogal")
    case _:
        print("não é uma vogal")
```
