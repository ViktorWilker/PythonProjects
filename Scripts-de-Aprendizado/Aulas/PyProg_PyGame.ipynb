```python
# DESENHANDO OBJETOS (NÃ£o esquecer: TERMINAL> pip install pygame ou ALT+Enter)

import pygame
from pygame.locals import *
from sys import exit

pygame.init()

largura = 1080
altura = 720

tela = pygame.display.set_mode((largura, altura))
pygame.display.set_caption("Joguinho test")

while True:
    tela.fill((255, 255, 255))

    # Chamando eventos de teclado ou sistema
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            exit()

    # Desenhando objetos
    pygame.draw.rect(tela, (255, 0, 0), (200, 20, 40, 50))
    pygame.draw.circle(tela, (0, 255, 0), (350, 250), 20)
    pygame.draw.line(tela, (255, 0, 255), (0, 100), (1080, 100), 10)

    pygame.display.update()
```

```python
# MOVIMENTANDO OBJETOS PELO TECLADO

import pygame
from pygame.locals import *
from sys import exit

pygame.init()

largura = 640
altura = 480

x = largura/2
y = altura/2

tela = pygame.display.set_mode((largura, altura))
pygame.display.set_caption('O Jogo...')
relogio = pygame.time.Clock()

while True:
  relogio.tick(30)
  tela.fill((0, 0, 0))
  # Chamando eventos de teclado ou sistema
  for event in pygame.event.get():
    if event.type == QUIT:
      pygame.quit()
      exit()

    if event.type == KEYDOWN:
      if event.key == K_a:
        x-=20
      if event.key == K_d:
        x+=20
      if event.key == K_w:
        y-=20
      if event.key == K_s:
        y+=20

  pygame.draw.rect(tela, (255, 0, 0), (x, y, 40, 50))

  pygame.display.update()
```
