```python
#BLOCO IF
x = 15
if x > 5:
    print("é maior que 5")
    if x < 8:
        print("também é menor que 8")
if x == 10:
    print("x vale 10!")
```

```python
# BLOCO IF ELSE
if x != 8:
    print("é diferente de 8")
else:
    print("é igual a 8")
```

```python
 #BLOCO Elif
if x <= 10:
    print("frio")
elif x <= 20:
    print("morno")
elif x <= 30:
    print("quente")
else:
    print("INFERNO")
```

```python
#BLOCO COM AND
if x >= 10 and x <= 20:
    print("x está entre 10 e 20")
else:
    print("Não está entre 10 e 20")
```

```python
#BLOCO COM OR
if x >= 10 or x <= 20:
    print("x é maior ou igual que 10 ou menor ou igual que 20")
else:
    print("x não é maior ou igual a 10 ou não é menor ou igual a 20")
```

```python
# BLOCO COM IN
escola = "SENAI Ary Torres"

if 'Ary' in escola:
    print("Tem Ary")
else:
    print("Não tem Ary")
```

```python
# BLOCO COM NOT IN
escola = "SENAI Ary Torres"

if 'Ary' not in escola:
    print("Não tem Ary")
else:
    print("tem Ary")
```

```python
# DESAFIO 05
"""
Escreva um programa que leia a velocidade de um carro.

Se ele ultrapassar 80km/h, mostre uma mensagem dizendo que
ele foi multado.

A multa custará R$ 7,00 por cada Km acima do limite.
"""

vel = float(input("Velocidade captada: "))
velMax = 80
multa = 7.00

if vel > velMax:
    print(f"Sua Multa é de {multa * (vel - velMax):,.2f}R$")
else:
    print("Velocidade dentro do limite")
```

```python
# DESAFIO 10
"""
Faça um programa que leia três números e mostre qual é o
maior e menor.
"""

n = input("Insira 3 números separados por espaço: ").split()
x = int(n[0])
y = int(n[1])
z = int(n[2])

if x > y:
    if x > z:
        print(f"O maior número é: {x}")

if y > x:
    if y > z:
        print(f"O maior número é: {y}")

if z > x:
    if z > y:
        print(f"O maior número é: {z}")

if x < y:
    if x < z:
        print(f"O menor número é:  {x}")

if y < x:
    if y < z:
        print(f"O menor número é: {y}")

if z < x:
    if z < y:
        print(f"O menor número é: {z}")
```

```python
# DESAFIO 12
"""
Escreva um programa para aprovar um empréstimo bancário
para a compra de uma casa. O programa vai perguntar o valor
da casa, o salário do comprador e em quantos anos ele vai
pagar.

Calcule o valor da prestação mensal, sabendo que ela não
pode exceder 30% do salário ou então o empréstimo será
negado.

"""

print("Responda as seguintes questões:")

value = float(input("\nQual o Valor da casa: "))
sal = float(input("Qual seu salário atual: "))
temp = float(input("Em Quantos anos será quitado: "))

taxa = 30/100
meses = temp * 12
prestacao = value / meses

if prestacao <= sal * taxa:
    print(f"\nNão poderá obter um empréstimo, \nprestação: {prestacao:,.3f} R$ \n para: {sal * taxa}")


else:
    print(f"Empréstimo aprovado: \nprestação: {prestacao:,.3f} R$ \n para: {sal * taxa}")
```

```python
# DESAFIO 13
"""
Escreva um programa que leia um número inteiro e peça para o
usuário escolher qual será a base de conversão:

1 para binário bin()
2 para octal oct()
3 para hexadecimal hex()
"""

n = int(input("Insira um numero inteiro: "))
opc = int(input("Qual será a base de conversão? \n1) Binário \n2) Octal \n3) Hexadecimal \n"))

if opc == 1:
    print(f"O inteiro {n} em Binário: {bin(n)}")
elif opc == 2:
    print(f"O inteiro {n} em Octal: {oct(n)}")
else:
    print(f"O inteiro {n} em Hexadecimal: {hex(n)}")
```

```python
# DESAFIO 15
"""
Faça um programa que leia o ano de nascimento de um jovem
e informe, de acordo com sua idade:

Se ele ainda vai se alistar ao serviço militar
Se é a hora de se alistar
Se já passou o tempo do alistamento

Seu programa também deverá mostrar o tempo que falta ou
que passou do prazo.
"""

ano = int(input("\nInsira um ano de nascimento: "))

idade = 2024 - ano

if idade == 18:
    print("Deve se alistar")
elif idade > 18:
    print(f"Já passou o tempo de alistar-se por {idade - 18} anos")
else:
    print(f"Ainda vai se alistar em {18 - idade} anos")
```

```python
# DESAFIO 16
"""
Crie um programa que leia duas notas entre 0 a 10 de um aluno
e calcule sua média, mostrando uma mensagem no final, de
acordo com a média atingida.

Média abaixo de 5.0: REPROVADO
Média entre 5.0 a 6.9: RECUPERAÇÃO
Média igual ou superior a 7.0: APROVADO
"""

nota1 = float(input("\nInsira a 1° nota: "))
nota2 = float(input("Insira a 2° nota: "))

média = (nota1 + nota2) / 2

if média >= 7.0:
    print("Aprovado")
elif média >= 5.0 and média <= 6.9:
    print("Recuperação")
else:
    print("Reprovado")
```

```python
# DESAFIO 17
"""
A confederação Nacional de Natação precisa de uma programa
que leia o ano de nascimento de uma atleta e mostre sua
categoria, de acordo com a idade.

Até 9 anos: MIRIM
Até 14 anos: INFANTIL
Até 19 anos: JUNIOR
Até 24 anos: SÊNIOR
Acima: MASTER
"""

ano = 2024
Nasci = int(input("Insira o ano de nascimento: "))

if ano - Nasci <= 9:
    print("Mirim")

elif ano - Nasci <= 14:
    print("Infantil")

elif ano - Nasci <= 19:
    print("Junior")

elif ano - Nasci <= 24:
    print("Sênior")
else:
    print("Master")
```
