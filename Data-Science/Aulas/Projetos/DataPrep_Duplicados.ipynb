# Preparação de Dados (Data Prep)
### Tratando Dados duplicados

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        "aluno": ["João", "Maria", "Helena", "Joel", "Gabriel"],
        "idade": [30, 25, 18, 29, 45]
    }
)

df
```

```python
df['idade'].duplicated().sum()
```

```python
df.duplicated().sum()
```

```python
#concatenação de dataFrame
df_dup = pd.concat((df, df))
df_dup.sort_values(by = 'aluno')
```

```python
df_dup.duplicated().sum()
```

```python
df_dup['aluno'].duplicated()
```

```python
df_unicos = df_dup.drop_duplicates()
df_unicos
```
