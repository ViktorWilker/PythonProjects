```python
from google.colab import output
import pandas as pd
import time

path = "/content/drive/MyDrive/Colab Notebooks/planilhas/"
GovernosXLSX = []

with open("/content/drive/MyDrive/Colab Notebooks/planilhas/GovernoPlaninhas.txt", 'r') as arquivo:
    Lista = arquivo.readlines()
    for i in Lista:
        GovernosXLSX.append(i.strip('\n'))

planilhas = []

for i in range(len(GovernosXLSX)):
    currentFile = pd.read_excel(f"{path}{GovernosXLSX[i]}" , None)
    planilhas.append(currentFile)


for i in range(len(planilhas)):
    output.clear()
    print(f"Iniciando planinha: {GovernosXLSX[i]}")
    time.sleep(3)
    print(planilhas[i])
    time.sleep(3)
```
