# Data Prep


```python
#convertendo os tipos
df.Preço = df.Preço.replace("'", "", regex=True)

df = df.astype(
    {
        'Preço': "float"
    }
)

df.Data_Venda = pd.to_datetime(df.Data_Venda, format = "%Y-%m-%d")
```
