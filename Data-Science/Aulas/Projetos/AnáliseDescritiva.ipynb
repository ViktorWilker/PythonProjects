Para obter uma visão clara dos dados.

```python
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/base_dados_vendas_tratada.csv'
df = pd.read_csv(file_path, parse_dates=['dt_venda', 'dt_entrega'], index_col = 'dt_venda')
```

```python
#count - conta a quantidade de valores não nulos de uma coluna

df.count()
df.dt_entrega.count()
```

```python
# sum - somar os valores de linhas ou colunas

colunas = ["preco_custo", "quantidade", "valor_final"]
df[colunas].sum()
```

```python
#média
df[colunas].mean().round(2)
```

```python
#max

df[colunas].max()
```

```python
#min

df[colunas].min()
```

```python
#mediana

df[colunas].median()
```

```python
#moda

df[colunas].mode()
```

```python
#desvio padrão

df[colunas].std().round(2)
```

```python
#Quantis calcula os percentis
df[colunas].quantile(0.75)
df[colunas].quantile(0.25)
```

```python
df['descricao_produto'].value_counts().sort_index()
```

## Filtrando os dados

```python
#Testes Lógicos
#Exemplo 01
filtro_vf = df['valor_final'] < 1000
df[filtro_vf]
```

```python
# Exemplo 02
filtro_vf = df['valor_final'] < 1000
filtro_loja = df['nome_loja'] == "Filial SP"

df_valor_loja = df[filtro_vf & filtro_loja]
```

```python
df_valor_loja[colunas]
```

```python
# Filtrando os dados com o método query
df.query("nome_loja == 'Filial SP' and valor_final < 1000")
```

```python
df_Matriz_Faturamento = df.query("descricao_produto == 'Fritadeira elétrica' or nome_loja == 'Loja Matriz'")
df_Matriz_Faturamento[colunas].sum().round(2)
```

# Agrupando dados com o método groupby()

```python
# Exemplo 01
df.groupby('nome_funcionario')['valor_final'].sum()
```

```python
# Exemplo 02
df.groupby(['nome_funcionario', 'descricao_produto'])[['preco_custo', 'quantidade', 'valor_final']].sum()
```

```python
df.groupby(['categoria','descricao_produto']).agg(
    {
        'quantidade': ['max', 'min'],
        'preco_custo': 'mean',
        'valor_final': 'sum'
    }
)
```
