# Transformar Colunas em Linhas no DataFrame
### Vamos criar um df com metas mensais para diferentes funcionários

```python
import pandas as pd
```

```python
# DataFrame com exemplos de metas
data = {
    'nome_funcionario': ['João', 'Maria', 'Pedro'],
    '01/01/2024': [5000, 7000, 6000],
    '01/02/2024': [6000, 8000, 5000],
    '01/03/2024': [5500, 7500, 6500]
}

df = pd.DataFrame(data)
df
```

```python
# Transformar Colunas em Linhas - Método melt()
df.melt(id_vars=['nome_funcionario'],
    var_name='dt_meta',
    value_name= 'vlr_meta')
```
