```python
df_preju = df.query("Lucro < 0")
```

```python
df_preju.groupby(['País']).agg(
    {
        'Valor Total c/ Desconto': 'sum',
        'Custo Total': 'sum',
        'Lucro': 'sum',
    }
).round(2).sort_index()
```

```python
df_preju.groupby(['Segmento']).agg(
    {
        'Valor Total c/ Desconto': 'sum',
        'Custo Total': 'sum',
        'Lucro': 'sum'
    }
)
```

```python
df_preju.groupby(['Produto']).agg(
        {
        'Valor Total c/ Desconto': 'sum',
        'Custo Total': 'sum',
        'Lucro': 'sum'
    }
).round(2)
```

```python
df_preju.groupby('Ano').agg(
    {
        'Valor Total c/ Desconto': 'sum',
        'Custo Total': 'sum',
        'Lucro': 'sum'
    }
).round(2)
```

## Prejuízos
- País com mais prejuízo: EUA
- Segmento com mais prejuízo: Grandes Empresas
- Produto com maior prejuízo: Produto 1
- Ano com maior prejuízo: 2019
