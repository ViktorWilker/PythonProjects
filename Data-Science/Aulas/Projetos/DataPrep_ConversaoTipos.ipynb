# Preparação dos Dados (Data Prep)
Manipulações de dados nas estapas **iniciais de projetos de dados**

### Conversões de tipos de dados

```python
import pandas as pd
```

```python
datas = {
    'variavel_a': ['1,1','2,2','3,3'],
    'variavel_b': ['4', '5', '6'],
    'variavel_c': ['10/10/2021', '11/12/2020', '15/09/2019']
}

df = pd.DataFrame(datas)
df.head()
```

```python
df.info()
```

```python
df.dtypes
```

```python
df.head()
```

```python
df.tail()
```

```python
df['variavel_a'] = df['variavel_a'].replace(',', '.', regex=True)
df.head()
```

```python
# O método astype converte o tipo de variávies
# De acordo com o que for especificado no dicionário

df = df.astype(
    {
        'variavel_a': 'float64',
        'variavel_b': 'int64'
    }
)
```

```python
df.dtypes
```

```python
# Formatando datas
df['variavel_c'] = pd.to_datetime(df['variavel_c'], format = '%d/%m/%Y')
df
```
